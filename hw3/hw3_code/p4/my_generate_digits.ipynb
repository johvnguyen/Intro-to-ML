Plotting the first 100 digits in a 10 x 10 grid  
This is just to remind ourselves what the data looks like

```python
def plot_digits(data):
    fig, ax = plt.subplots(10, 10, figsize=(8, 8),
                           subplot_kw=dict(xticks=[], yticks=[]))
    fig.subplots_adjust(hspace=0.05, wspace=0.05)
    for i, axi in enumerate(ax.flat):
        im = axi.imshow(data[i].reshape(8, 8), cmap='binary')
        im.set_clim(0, 16)
```

```python
plot_digits(digits.data)
```

```python
def myPCA(X, alpha):
    n_data, n_features = X.shape
    mean_vec = []
    
    # Calculating the mean vector
    for i in range(n_features):
        i_mean = np.mean(X[:,[i]])
        mean_vec.append([i_mean])
    
    mean_vec = np.array(mean_vec)
    
    # Computing the covariance matrix
    cov_mat = np.zeros((n_features, n_features))
    
    for i in range(n_data):
        x_vec = X[i]
        #print(x_vec)
        diff = x_vec - mean_vec.T
        diff = np.array(diff)
        prod = np.matmul(diff.T,diff) # We transpose the first entry because we are working with row vectors
        
        cov_mat += prod
        
    cov_mat = (1/n_data) * cov_mat
    cov_mat = np.array(cov_mat)
    
    # Computing eigenvalues
    cov_eig_val, cov_eig_vec = np.linalg.eig(cov_mat)
    
    # np.linalg.eig returns eigenvectors which match the columns; we are working with row vectors
    cov_eig_vec = cov_eig_vec.T
    cov_eig_val = cov_eig_val
    
    # Sorting eigenvalues from greatest to least
    index_array = np.argsort(cov_eig_val)
    cov_eig_val_ = []
    cov_eig_vec_ = []
    
    # Eigenvalues may not be sorted; they are in order of multiplicity.
    # Sort the eigenvalues in descending order
    for i in range(len(index_array)):
        j = index_array[i]
        
        cov_eig_val_.insert(0,cov_eig_val[j])
        cov_eig_vec_.insert(0,cov_eig_vec[j])
        
    cov_eig_val = np.array(cov_eig_val_)
    cov_eig_vec = np.array(cov_eig_vec_)
    
    # Calculate how many eigenvalues/eigenvectors to reduce to
    extracted_eig_val = []
    extracted_variance = 0
    total_variance = cov_eig_val.sum()
    trans = []
    d = 0
    
    while((extracted_variance / total_variance) * 100 < alpha):
        d += 1
        component = cov_eig_val[0]
        cov_eig_val = np.delete(cov_eig_val, 0)
        
        component_vec = cov_eig_vec[0]
        cov_eig_vec = np.delete(cov_eig_vec, 0, axis = 0)
        extracted_eig_val.append(component)
        
        trans.append(component_vec)
        
        extracted_variance = sum(extracted_eig_val)
        
    trans = np.array(trans)
    trans = trans.T
    
    return d, trans, mean_vec
```

We are using sklearn's PCA here, these will be replaced by your code for PCA2 (HW3, Q4)  
We are creating a low-dimensional projection so that 90% of the variance is preserved after projection  
The exact dimensionality needed can be determined using the eigen-values of the covariance matrix of the data  
For preserving 90% of the variance, we need 21 dimensions, original data had 64 dimensions! 

```python
d_90, trans_90, mean_vec_90 = myPCA(digits.data, 90)
trans = np.array(trans)

data90 = digits.data @ trans_90
print(data90.shape)
```

Next we fit a mixture of Gaussians on the projected data  
The first step is to figure how many components to use  
We use a model selection criterion called AIC, we have not discussed this in class  
Just use this bit as is

```python
n_components = np.arange(50, 510, 20)
models = [GaussianMixture(n, covariance_type='full', random_state=0)
          for n in n_components]
aics = [model.fit(data90).aic(data90) for model in models]
plt.plot(n_components, aics);
```

Looks like 250 components would suffice  
So we will fit a mixture of 250 Gaussians to the projected data  
Note that we are doing the fit in a 21 dimensional space

```python
gmm90 = GaussianMixture(250, covariance_type='full', random_state=0)
gmm90.fit(data90)
print(gmm90.converged_)
```

Now, we will sample 100 samples from the learned mixture of 250 Gaussians  
Each sample will be a point in the 21 dimensional (projected) space  

```python
data_new90 = gmm90.sample(100)[0]
data_new90.shape
```

Finally, we will transform these 100 points in 21 dimensions back to the original 64 dimensional space  
We can then visualize them, and see if they look like digits from the digits dataset

```python
digits_new90 = data_new90 @ trans_90.T
plot_digits(digits_new90)
```
